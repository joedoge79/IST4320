```python
import tkinter as tk
from tkinter import messagebox
import sqlite3

# Connect to SQLITE
sql_connect = sqlite3.connect("database.db")
cur = sql_connect.cursor()

# Creates table
cur.execute("""
CREATE TABLE IF NOT EXISTS app_data (
    principal REAL,
    interest_rate REAL,
    years INTEGER,
    final_amount REAL
)
""")

sql_connect.commit()

def calculate_and_save():
    try:
        principal = float(principal_entry.get())
        rate = float(rate_entry.get())
        years = 5

        final_amount = principal * (1 + rate / 100) ** years

        query = """
        INSERT INTO app_data (principal, interest_rate, years, final_amount)
        VALUES (?, ?, ?, ?)
        """

        cur.execute(query, (principal, rate, years, final_amount))
        sql_connect.commit()

        result_label.config(
            text=f"After 5 years, you can expect: ${final_amount:,.2f}"
        )

    except ValueError:
        messagebox.showwarning("Invalid Input", "Please enter valid numbers.")

def clear_fields():
    principal_entry.delete(0, tk.END)
    rate_entry.delete(0, tk.END)
    result_label.config(text="")

# Tkinter popup
window = tk.Tk()
window.title("Compound Interest Calculator")
window.geometry("800x800")

title_label = tk.Label(
    window,
    text="For those who invest a principal amount into a stock and want to see how much they can expect in 5 years, highly recommended to start investing ASAP",
    font=("Arial", 12, "bold"),
    wraplength=300,   # 👈 THIS controls when text wraps
    justify="center"  # optional: center align text
)
title_label.pack(pady=10)

principal_label = tk.Label(window, text="Enter principal investment:")
principal_label.pack()

principal_entry = tk.Entry(window, width=45)
principal_entry.pack(pady=5)

rate_label = tk.Label(window, text="Enter compound interest rate (%):")
rate_label.pack()

rate_entry = tk.Entry(window, width=45)
rate_entry.pack(pady=5)

submit_button = tk.Button(
    window,
    text="Calculate and Save",
    bg="black",
    fg="white",
    width=24,
    command=calculate_and_save
)
submit_button.pack(pady=5)

clear_button = tk.Button(
    window,
    text="Clear",
    bg="black",
    fg="white",
    width=24,
    command=clear_fields
)
clear_button.pack(pady=5)

result_label = tk.Label(window, text="", fg="purple")
result_label.pack(pady=10)

window.mainloop()

sql_connect.close()
```
